```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('weather.csv')
series = pd.Series(df['MinTemp'])
sns.distplot(series)              
```

```python
sns.distplot(series, kde=False, rug=True)
```

```python
series = pd.Series(df['MaxTemp'])
sns.kdeplot(series, shade=True)
```

```python
series = pd.Series(df['MaxTemp'])
sns.distplot(series, hist=False, rug=True)
```

```python
df = pd.DataFrame(df, columns=['Rainfall', 'Evaporation'])
sns.jointplot(x = 'Rainfall', y = 'Evaporation', data = df)
```
