##UNIVARIATE Analysis
uni variate analysis refers to analysis of any single variable at a time.

It also help in further analysis of bivariate and multivariate.

It helps us to find out the outliers.

###Volatile Acid Count plot

```python
plt.hist(x='volatile.acidity',data=data,bins=50)
plt.grid()
plt.title('Count plot of Volatile acidity in white wine')
```

###Analysis of count plot of Volatile Acidity
1. There are many  outliers after 0.6
2. Major sample of wine have the volatile acidity content between 0.2 to  0.4
3. It is Normal Distribution which is seemed to be skewed.



##Fixed Acidity count plot

```python
plt.hist(x='fixed.acidity',data=data,bins=50)
plt.grid()
plt.title('Count plot of Fixed acidity in white wine')
```

##Analysis of Count Plot of Fixed Acidity
1. Major sample of wine are in the range of 6 to 8
2. There are very few outlier which are in the range of 10 - 12
3. It is normal distribution

#Bivariant analysis

```python
plt.figure(figsize=(12,8))
sns.heatmap(data.corr(),annot=True)
```

this is correlation graph between the different variables of the wine
##Quality vs Alcohol

```python
plt.figure(figsize=(10,6))
sns.regplot(x='quality', y='alcohol', data=data, x_jitter=0.3, scatter_kws={'alpha':0.2})
sns.despine()
```

1. Good quality wines sample are less in number and have high concentration
2. Mjor sample have quality grade of 5,6,7
3. Major sample have concentration in range of 9-11 for quality grade 5, 9-12 for quality grade 6 and 9-13 for quality grade 7
##Quality vs Density

```python
sns.regplot(x='quality' , y='density', data=data, x_jitter=0.35, scatter_kws={'alpha':0.3})
```

1. Quality vs Density have correlation coefficient of -0.31 
2. Top rated wine samples (quality rating of 9) are observed to have lower densities of around 0.99 
3. Average quality wine samples have densities between 0.99-1.00 .
##Alcohol vs Density

```python
plt.figure(figsize=(10,6))
sns.regplot(x='alcohol', y='density', data=data, x_jitter=0.2, scatter_kws={'alpha':0.15})
sns.despine()
```

density and alcohol have correlation coefficient of -0.78

Increase in concentration leads to decrease in alcohol concentration

there are some outlier
##Residual sugar v/s Quality

```python
sns.regplot(x='quality',y='residual.sugar',data=data,x_jitter=0.2,scatter_kws={'alpha':0.2})
plt.title("Quality  and Residual Sugar regplot")
plt.show()
```

Residual sugar content of good quality wine is very less

correlation cofficient b/w these is -0.098.

we can also notice that average quality wines (5 and 6) has residual sugar contents between 0-20
##pH v/s Quality

```python
sns.boxplot(x='quality',y='pH',data=data)
plt.title("Quality and pH Boxplot")
plt.show()
```

correlation cofficient b/w these is 0.099.

As we know that pH's for white wine always lies between 3.0 to 3.4 .

As we can observe that quality of wines would be better when the pH's falls between 3.1 to 3.3.
##Density v/s Quality

```python
sns.pointplot(x='quality',y='density',data=data)
plt.title("Quality and Density pointplot")
plt.show()
```

As density increases, perceived quality decreases

corelation cofficient b/w this is -0.31.

##Sulphates v/s Quality

```python
sns.boxplot(x='quality',y='sulphates',data=data)
plt.title("Quality and sulphates Boxplot")
plt.show()
```

the correlation between these two is 0.054

As we can notice that the sulphates amount lies between 0.4 and 0.6 in wines.

As sulphates increase, so does quality increases
#Mutivariate Analysis
