```python
# Question 1

import numpy as np
arr=np.random.random((10,1))
print(arr)
print('*'*40)
mean=np.mean(arr)
print(mean)
```

```python
# Question 2
arr2=np.random.random((20,1))
print(arr2)
print('*'*40)
std_dev=np.std(arr2)
print(std_dev)
var=std_dev**2
print('*'*40)
print(var)
```

```python
# Question 3

a=np.random.random((10,20))
print(a)
print('*'*40)
b=np.random.random((20,25))
print(b)
print('*'*40)
c=np.matmul(a,b)
print(c)
print('*'*40)
sum=np.sum(c)
print(sum)
```

```python
# Question 4

A=np.random.random((10,1))
print(A)
print('*'*40)

def func(x):
    return (1/(1+np.exp(-x)))

ans=np.apply_along_axis(func,0,A)
print(ans)
```
